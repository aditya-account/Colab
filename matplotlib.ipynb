#plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,4,12,3,4,1]
plt.plot(x,y,'b',linewidth = 3.0)
plt.ylabel('number*2')
plt.xlabel('number')
plt.title('Plot')
plt.show()
```

#bar graph

```python
number_of_people = [1,2,3,4,5,6,7]
age = [34,36,70,65,35,13,52]
plt.bar(age,number_of_people,label = "Example 1",color = 'r')
plt.legend()
plt.xlabel("age")
plt.ylabel("number_of_people")
plt.title("Bar graph")
plt.show()
```

```python

import numpy as np
x = [1,2,3,4]
x = np.sin(x)
print(x)
```

#histogram graph

```python
student_marks = np.random.randint(0,100,20)
age = np.random.randint(30,200,2)
plt.hist(student_marks,color = 'c',rwidth = 6)
plt.xlabel("age")
plt.ylabel("student_marks")
plt.title("Histogram")
plt.show()
```

#scattered graph

```python
number_of_people = [1,2,3,4,5,6,7]
age = [34,36,70,65,35,13,52]
plt.scatter(age,number_of_people,label = "Example 1",color = 'r')
plt.legend()
plt.grid()
plt.xlabel("age")
plt.ylabel("number_of_people")
plt.title("scatter graph")
plt.show()
```

#plot - randint(start,end,step)

```python
number_of_people = np.random.randint(0,100,10)
age = np.random.randint(0,50,10)
plt.plot(age,number_of_people,label = "Example 1",color = 'r')
plt.legend()
plt.xlabel("age")
plt.ylabel("number_of_people")
plt.title("plot")
plt.show()
```

#pie chart

```python
cities = ["hyderabad","delhi","agra","gujarat"]
people_like = [10,12,8,15]

fig = plt.figure(figsize = (10,5))
plt.pie(people_like, labels = cities,explode = [0.1,0.1,0.1,0.2])
plt.title("pie chart")
plt.show()
```

```python
x1 = [2,4,6,8,10,16]
y1 = [1,3,5,7,9,25]

plt.plot(x1,y1,label = "line1",color = "r")

x2 = [10,43,23,15,10]
y2 = [7,6,13,24,17]

plt.plot(x2,y2,label = "line2",color = "g")
plt.grid(color = "c",linestyle = "--")
plt.legend()
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,4,6,8,10,14]

plt.plot(x,y,color = "green", linestyle = "--",linewidth = 3, marker = '*', markerfacecolor = 'red', markersize = 15)

plt.xlim(1,8)
plt.ylim(1,8)

plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

#sub plot

```python
x1 = [1,2,3,4,5,6]
y1 = [2,4,6,8,10,14]

x2 = [10,43,23,15,10]
y2 = [7,6,13,24,17]

x3 = [2,4,6,8,10]
y3 = [3,6,9,10,12]

plt.subplot(2,3,1)
plt.plot(x1,y1,'b*--')

plt.subplot(2,3,2)
plt.plot(x2,y2,'b*--')

plt.subplot(2,3,3)
plt.plot(x3,y3,'b*--')
plt.show()
```

#subplots with different graphs

```python
x1 = [1,2,3,4,5,6]
y1 = [2,4,6,8,10,14]

x2 = [10,43,23,15,10]
y2 = [7,6,13,24,17]

x3 = [2,4,6,8,10]
y3 = [3,6,9,10,12]

x4 = ["earth","mars","venus","jupiter"]
y4 = [7,6,13,24]

plt.subplot(2,3,1)
plt.plot(x1,y1,'b*--')

plt.subplot(2,3,2)
plt.scatter(x2,y2)

plt.subplot(2,3,4)
plt.bar(x3,y3)

plt.subplot(2,3,5)
plt.pie(y4,labels=x4)


plt.show()
```
